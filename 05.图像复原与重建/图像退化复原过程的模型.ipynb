```python
import numpy as np
import cv2
import matplotlib
import matplotlib.pyplot as plt #最后两个库为了画子图使用
```

```python
def draw_gaussian(x, mu, sigma):
    return np.exp(-1*(((x - mu)**2) / (2*sigma**2))) / (np.sqrt(2*np.pi) * sigma)

#高斯噪声是指它的概率密度函数服从高斯分布（即正态分布）的一类噪声。
def addGaussianNoise(image, G_NoiseNum):
    # 不copy原图会被污染
    G_Noiseimg = image.copy()
    # 噪点数量
    for i in range(G_NoiseNum):
        # 随机生成坐标
        temp_x = np.random.randint(0, G_Noiseimg.shape[0])
        temp_y = np.random.randint(0, G_Noiseimg.shape[1])
        G_Noiseimg[temp_x][temp_y] = 255
    return G_Noiseimg

# 图像转化成灰度图像
img = cv2.imread('test.png', cv2.IMREAD_GRAYSCALE)
G_Noiseimg = addGaussianNoise(img, 10000)

titles = ['original_img', 'G_Noiseimg']
images = [img, G_Noiseimg]

plt.figure(dpi=250) #指定输出像素大小
for i in range(2):
    plt.subplot(1, 3, i + 1)
    plt.imshow(images[i])
    plt.title(titles[i], fontsize=8)
    plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#椒盐噪声也称泊松噪声。，一般是由图像传感器、传输信道及解码处理等产生的黑白相见的亮暗点噪声，椒盐噪声常由图像切割产生。
#椒盐噪声是指两种噪声： 椒盐噪声 = 椒噪声 + 盐噪声。盐噪声一般是白色噪声，椒噪声一般是黑色噪声，前者高灰度噪声，后者属于低灰度噪声，一般两种噪声同时出现
#呈现在图像上就是黑白杂点。
#生成椒盐噪声的步骤如下：首先读取图片，然后确定噪声点的数量，最后随机生成噪声点在图中的坐标，随机定义01生成器
#当结果为0代表生成椒噪声，结果为1表示盐噪声，代码如下：

def salt_pepperNoise(img, noiseNum):
    image = img.copy()
    ndim = img.ndim
    row, col = img.shape[0:2]
    for i in range(noiseNum):
        x = np.random.randint(0, row)
        y = np.random.randint(0, col)
        # 0-salt, 1-pepper
        salt_or_pepper = np.random.randint(0, 2)
        if salt_or_pepper == 0:
            image[x, y] = 0
        else:
            image[x, y] = 255
    return image
#实验结果如下所示，和上述高斯噪声不一样的地方在于高斯噪声读取的图片为灰度图，此次的椒盐噪声用的则是RGB的彩色图像，
#代码上就是image[x, y, :]和image[x, y]的区别，彩色图像多一个维度。

# 图像转化成灰度图像
img = cv2.imread('test.png', cv2.IMREAD_GRAYSCALE)
salt_pepperNoise = salt_pepperNoise(img, 10000)

titles = ['original_img', 'salt_pepperNoise']
images = [img, salt_pepperNoise]

plt.figure(dpi=250) #指定输出像素大小
for i in range(2):
    plt.subplot(1, 3, i + 1)
    plt.imshow(images[i])
    plt.title(titles[i], fontsize=8)
    plt.xticks([]), plt.yticks([])
plt.show()
```
