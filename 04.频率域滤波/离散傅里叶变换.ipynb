```python
import math
import numpy as np
import matplotlib.pyplot as plt
T=32

pi = math.pi
N = 1024
W = np.exp((2*pi/N)*1j)
nature = np.arange(N)
DFT_Matrix = np.ones([N,N],np.complex_)
for row in range(N):
    DFT_Matrix[row]=W**(-nature*(row))
```

```python
ls =  [1 if num%T<T/2 else 3 for num in range(N) ] 
#ls =  [1,3,1,3]
ls
```

```python
DFT_Matrix
```

```python
y=np.array(ls)
Y = np.matmul(DFT_Matrix, y)
y_idx = np.linspace(0, 2 * pi, N)
plt.plot(y_idx, np.absolute(Y))
```
